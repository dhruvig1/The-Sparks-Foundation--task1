Task: Predicting the percentage of a student on the basis of the no. of hours studied. 
Prediction using Supervised ML

It is a simple linear regression problem.

#### IMPORTING MODULES
Importing necessary modules:

```python
import pandas as pd #to read the file
import numpy as np 
import matplotlib.pyplot as plt
import csv
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
```

#### IMPORTING DATASET
Next, we import the dataset using pandas. The dataset has two columns. It is read and analysed.

```python
%matplotlib inline
data="https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df=pd.read_csv(data)
df.set_index(['Hours','Scores'])
x=df['Hours']
y=df['Scores']
print(df)
```

#### PLOTTING SCATTER DIAGRAM
Analysing, we can say that scores are dependent on number of hours of study.
Both, hours and scores, are discrete values. To find out the relationship between the two, we plot a scatter diagram.

```python
df.plot()
#colors=np.linespace(0,100,25)
df.plot(kind='scatter' , x='Hours', y='Scores')
```

By looking at the scatter plot, we can easily say that the data is positively correlated as the datapoints are closely spaced.
A correlation between x and y(in this case Hours and Scores) is said to be positive if the correlation coefficient lies between 0.8 and 1.

#### FINDING CORRELATION COEFFICIENT
We find correlation coefficient.

```python
corcoeff=df.corr(method='pearson')['Hours']['Scores']
print('Karl Pearson"s Correlation Coefficient= ', corcoeff)
```

#### TRAIN AND TEST DATA
Now we use the divide the data into train and test sets and we use Scikit-Learn's train_test_split and train the algorithm.

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
lin = LinearRegression()
lin.fit(x_train.values.reshape(-1,1),y_train) 
print('via sklearn model: ')
print('The slope is {0} and the intercept is {1}. Formula is y={0:.4f}x+{1:.4f}'.format(lin.coef_[0],lin.intercept_))
```

#### REGRESSION LINE
Regression line looks like:

```python
line = lin.coef_*x+lin.intercept_

plt.scatter(x, y)
plt.plot(x, line);
plt.show()
```

#### MAKING PREDICTIONS
Thereafter, we make predictions: 

```python
predicted_val= lin.predict(x_test.values.reshape(-1,1))
print('By Sklearn Model, R2 score is: ')
print(r2_score(y_test,predicted_val))
```

#### ESTIMATING SCORES
Since our linear regression model seems to be apt to estimate values of y from x. We substitute x=9.5 in the equation\n                                    y = 9.8416 * x + 2.6490\n
                           y = 9.8416 * 9.25 + 2.6490

```python
solution=lin.predict([[9.25]])
print('Predicted Score Of The Student Is: ')
print(solution[0])
```
